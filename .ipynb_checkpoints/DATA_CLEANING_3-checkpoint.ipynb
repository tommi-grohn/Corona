# Cleaning HSL's city bike traffic data into a dataframe and plotting a few time series of relevant measure points

## 1. A brief introduction

### Important notions


- In this notebook HSL city bike data is collected from Helsinki and Espoo


- A good overview how the stations are located in Helsinki and Espoo: https://kaupunkipyorat.hsl.fi/en/helsinki/stations


- The cleaned data is added to a dataframe which is exported

### Adding libraries

```python
import numpy as np
import pandas as pd
import datetime
import os
import matplotlib.pyplot as plt
```

### Manually controlled parameters

```python
# These regions will be plotted
station_ids_plotted = [4, 26, 113, 541, 729 ]
corresponding_station_names = ['Viiskulma', 'Kamppi (M)', 'Pasilan asema', 
                               'Aalto-yliopisto (M), Korkeakouluaukio', 'Leppävaaranaukio']
```

### The dataframe which will be exported as a csv

```python
# A dataframe sorted by countries

dtypes_hsl_traffic = np.dtype([
          ('Station id', int), # For example 40 corresponds a station in Hakaniemi 
          ('Station name', str),   # The name of the station
          ('Date', np.datetime64), # The current date
          ('Number of trips', float), # The number of people which starrted a trip from this station during the current date
          ])

data_hsl_traffic = np.empty(0, dtype=dtypes_hsl_traffic)
df_hsl_traffic = pd.DataFrame(data_hsl_traffic)   
```

## 2. Cleaning the data and adding it to a dataframe which will exported as a csv 

### Find out all the ids of stations and the corresponding station names

```python
# From this dataframe station ids and names are going to be read
data_frame_may = pd.read_csv('Data/HSL_data/2020-05.csv')

# Get station ids as list
station_ids = list(set(data_frame_may['Departure station id'].tolist()))

# And also matching station names to a list
corresponding_station_names = []

for current_id in station_ids:  
    corresponding_station_names.append(
        data_frame_may.loc[(data_frame_may['Departure station id'] == current_id), 'Departure station name'].iloc[0])
```

### Adding and cleaning Väylä's data to a dataframe


- HSL has divided its city bike data of different months in different csv-files


- Running next cell takes roughly 2-3 minutes

```python
# Loop over each file 
for current_file in os.listdir('Data/HSL_data'):
    
    # Informs the user which file is currently read
    print("Currently reading the file:")
    print(current_file)
    
    
    # Read the current file and add it to a temporary dataframe
    df_current_month = pd.read_csv('Data/HSL_data/' + current_file)[[
                      'Departure', 'Departure station id', 'Departure station name']]
    
    
    # Make sure the column types are the correct ones
    convert_dict = {
              'Departure': np.datetime64,  
              'Departure station id': int, 
              'Departure station name': str }
    df_current_month = df_current_month.astype(convert_dict) 

    
    # Get the day interval of current month, f.e. in June [1,2,..., 30]
    day_interval = range(df_current_month['Departure'].tolist()[-1].day,  
                         df_current_month['Departure'].tolist()[0].day, 1)
    len_day_interval = len(day_interval)
    
    # And the corresponding date interval
    first_date = df_current_month['Departure'].tolist()[-1]
    first_date = first_date.replace(hour=0, minute=0, second=0)
    date_interval = [first_date + datetime.timedelta(days=x) for x in range(len_day_interval)]
    
        
    # Loop over each station
    for i in range(len(station_ids)):
        
        current_id = station_ids[i]
        current_name = corresponding_station_names[i]
        
        df_current_month_by_id = df_current_month[(df_current_month['Departure station id'] == current_id)]
        
        # Loop over each day
        for j in range(len(day_interval)):
            
            # Get current day and date
            current_day = day_interval[j]
            current_date = date_interval[j]
            
            # Make a dataframe for the current day and id
            df_current_day_by_id = df_current_month_by_id[(df_current_month_by_id['Departure'].dt.day == current_day )]
            
            # Make a dataframe filtered by current day and station 
            number_trips = df_current_month_by_id[(df_current_month_by_id['Departure'].dt.day == current_day )].shape[0]
            
            # Now we can update one measure points total traffic of one day to the final dataframe
            df_new_row = pd.DataFrame({'Station id':[current_id], 'Station name': [current_name], 
                        'Date': [current_date], 'Number of trips': [number_trips] } )
            df_hsl_traffic = pd.concat([df_hsl_traffic, df_new_row])    
            
# Make the column 'Number of trips' to be int
df_hsl_traffic = df_hsl_traffic.astype({'Number of trips': int})
            
# Make the indexing to make more sense
df_hsl_traffic.reset_index(drop=True, inplace=True)            
```

### The dataframe is now fully implemented

```python
# Export the dataframe
output_file = os.path.join('Dataframes', 'df_hsl_traffic.csv')
df_hsl_traffic.to_csv(output_file, index=False)

df_hsl_traffic
```

## 3. Plotting a few time series of relevant measure points

```python
print('3 locations in Helsinki and 2 in Espoo will be plotted next. ')
print()
print()


# Date interval is for each measure point the same
date_interval = df_hsl_traffic.loc[(df_hsl_traffic['Station id'] == station_ids_plotted[0]), 
                    'Date'].tolist()

# Loop over measure points to be plotted
for i in range(len(station_ids_plotted)):
    
    # Current measure points and locations
    current_station_id = station_ids_plotted[i]
    current_station_name = corresponding_station_names[i]
    print('\033[1m' + 'Station id: ' + str(current_station_id) + ', Station name: ' + current_station_name)

    # Total traffic of the current measure point
    current_number_trips = df_hsl_traffic.loc[(df_hsl_traffic['Station id'] == station_ids_plotted[0]),
                    'Number of trips'].tolist()
    
    # Define the figure
    fig, host = plt.subplots(figsize=(26, 6))

    # Plot the traffic, infected and death data
    p1, = host.plot(date_interval, current_number_trips, marker = 'o',
                    linestyle='', color = "black", label="Number of trips in a day ()" )

    # Define the text for y-axis
    host.set_ylabel("Number of trips in a day ()")

    # Make little spikes for different y-axis
    host.tick_params(axis='y', colors=p1.get_color())
    host.tick_params(axis='x')
        
    plt.show()
```
